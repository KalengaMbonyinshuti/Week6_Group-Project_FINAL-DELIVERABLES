Check if there is no common songs between not hot and hot100.

```python
index_to_remove = []
n=0
for s,a in zip(nothot['title'], nothot['artist']):
    for sh, ah in zip(hot100['title'], hot100['artist']):
        if s == sh and a == ah: 
            index_to_remove.append(n)
    n += 1
print(index_to_remove)
```
